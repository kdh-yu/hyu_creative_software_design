```python
# Libraries
import random
from pprint import pprint
import time
import os
#from IPython import 
```

```python
# Class settings

class Player:
    def __init__(self, name, xpos=4, ypos=4):
        self.name = name
        self.pokemon = []
        self.xpos = xpos
        self.ypos = ypos
    
    def pokemon_add(self, pokemon):
        self.pokemon.append(pokemon)
    
    def fainted(self, pokemon):
        self.pokemon.remove(pokemon)

class Pokemon:
    def __init__(self, species, attribute, elemental_attack, hp, maxHP, strong_against, weak_against):
        self.species = species
        self.nickname = species
        self.attribute = attribute
        self.elemental_attack = elemental_attack
        self.physical_attack = 'Tackle'
        self.hp = hp
        self.maxHP = maxHP
        self.strong_against = strong_against
        self.weak_against = weak_against
    
    def set_nickname(self, nickname):
        self.nickname = nickname

Charmander = Pokemon('Charmander', 'Fire', 'Ember', 50, 50, 'Grass', 'Water')
Squirtle = Pokemon('Squirtle', 'Water', 'Water Gun', 50, 50, 'Fire', 'Grass')
Bulbasaur = Pokemon('Bulbasaur', 'Grass', 'Vine Whip', 50, 50, 'Water', 'Fire')
```

```python
# Battle Settings

def pokemon_encounter(wild):
    encountered = random.choices(wild, weights=[2,2,2,4])
    return encountered
    
def is_effective(attack, defend):
    weakenss_chart = {'Fire' : (1, 0.5, 2),
                    'Water' : (2, 1, 0.5),
                    'Grass' : (0.5, 2, 1)}
    defend_index = weakenss_chart.keys().index(defend)
    return weakenss_chart[attack][defend_index]

def hp_bar(now):
    left = '#' * round(now / 2)
    dmg = '─' * (25 - round(now / 2))
    return f'[{left}{dmg}]'
            
def move_pos(player, move):
    if move == 'n':
        if player.ypos > 1: 
            player.ypos += 1
    if move == 'e':
        if player.xpos < 77:
            player.xpos += 1
    if move == 'w':
        if player.xpos > 1:
            player.xpos -= 1
    if move == 's':
        if player.ypos < 9:
            player.ypos -= 1
```

```python
# UI

def battle_ui(player, enemy):
    print(f'''
┌─────────────────────────────────────────────────────────────────────────────┐
│                                    Enemy: {enemy.species}                                  │
│                                    {hp_bar(enemy.hp)}                                    │
│                                                                                    │
│                                                    1. Skill Attack                 │
│                                                    2. Physical Attack              │
│You: {player.pokemon[0].nickname}                                       3. Cure                         │
│{hp_bar(player.pokemon[0].hp)}                         4. Capture                                     │
│                                        5. Change Pokemon                           │
└─────────────────────────────────────────────────────────────────────────────┘''')

def show_map(player):
    ui_map = [['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  '],
            ['  ','  ','  ','  ','  ','  ','  ','  ','  ','  ']]
    xpos, ypos = player.xpos, player.ypos
    ui_map[ypos][xpos] = ' 𖨆'
    print('┌─────────────┐')
    for i in ui_map:
        print('│', end='')
        for j in i:
            print(j, end='')
        print('│')
    print('└─────────────┘')

def poke_appear():
    flash_1 = '''
################
################
################
################
################
################
################
################
################
    '''
    flash_2 = '''
    







    
    '''
    print(flash_1)
    time.sleep(1)
    print(flash_2)
    time.sleep(1)
    print(flash_1)
    time.sleep(1)
    print(flash_2)
    time.sleep(1)
    
```

```python
# Initial Settings
def set_player_name():
    name = input('what is your name : ')
    return name

def choose_pokemon(player):
    print("I'll give you a pokemon for a partner.")
    print('''
    1. Charmander
    2. Squirtle
    3. Bulbasaur''')
    while True:
        chosen_pokemon = int(input("What pokemon do you want? : "))

        if chosen_pokemon not in [1, 2, 3]:
            pass
        else:
            if chosen_pokemon == 1:
                player.pokemon_add(Charmander)
            elif chosen_pokemon == 2:
                player.pokemon_add(Squirtle)
            elif chosen_pokemon == 3:
                player.pokemon_add(Bulbasaur)
            break
        
    print(f"You recieved {player.pokemon[0].species}!")
    nickname_change_or_not = input("Will you give your pokemon a nickname? (yes / no)\n: ")

    if nickname_change_or_not == 'yes' or nickname_change_or_not == 'y':
        nickname = input('nickname : ')
        player.pokemon[0].set_nickname(nickname)
    else:
        pass
```

```python
# Battle Action

def skill_attack(player, enemy):
    print(f"{player.pokemon[0]} used {player.pokemon[0].elemental_attack}!")
    effective_multiplier = is_effective(player.pokemon[0].attribute, enemy.attribute)
    if effective_multiplier == 2:
        print("It's super effective!")
    elif effective_multiplier == 0.5:
        print("It's not very effective...")
    enemy.hp -= effective_multiplier * 10

def physical_attack(player, enemy):
    print(f"{player.pokemon[0]} used {player.pokemon[0].physical_attack}!")
    enemy.hp -= 9

def hp_restore(player):
    print(f"{player.name} used Full Restore!")
    print(f"{player.pokemon[0]}'s HP was restored.")
    player.pokemon[0].hp = player.pokemon[0].maxHP

def pokeball(player, enemy):
    if round(enemy.hp / enemy.maxHP) < 0.5:
        is_catched = random.choices(['success', 'fail'], weights=[90, 10])
    else:
        is_catched = random.choices(['success', 'fail'], weights=[15, 85])
    if is_catched == 'success':
        print("catched")

def change_pokemon(player):
    ind = 1
    for i in player.pokemon:
        print(f'{ind} : {i}')
        ind += 1
    to_what = int(input())
    if to_what == 0 or to_what < 0 or to_what > len(player.pokemon):
        pass  # Think more
    else:
        player.pokemon[0], player.pokemon[to_what - 1] = player.pokemon[to_what - 1], player.pokemon[0]

def action(act, player, enemy):
    if act == 1:
        skill_attack(player, enemy)
    if act == 2:
        physical_attack(player, enemy)
    if act == 3:
        hp_restore(player)
    if act == 4:
        pokeball(player, enemy)
    if act == 5:
        change_pokemon(player)
    else:
        physical_attack(player, enemy)

def enemy_action(player, enemy):
    eff = is_effective(enemy.attribute, player.pokemon[0].attribute)
    if eff == 0.5:
        print('평타')
        player.pokemon[0].hp -= 9
    else:
        print('스킬')
        player.pokemon[0].hp -= eff * 10
```

```python
# Play game!

# Wild Pokemon List
encounter_pokemon = [Charmander, Squirtle, Bulbasaur, None]

# Order 1
trainer_name = set_player_name()
player = Player(trainer_name)

# Order 2, 3
choose_pokemon(player)
step = 0

# Order 4~
while step <= 3:
    step += 1

    # Order 5
    show_map(player)
    direction = input("Which way? : ")
    move_pos(player, direction)
    
    # Order 6
    encounter = pokemon_encounter(encounter_pokemon)
    if encounter == None:  # Order 6 - A
        print("No pokemon appeared.")
    else:
        poke_appear()
        while True:
            battle_ui(player, encounter)
            act = int(input('What to do? : '))
            action(act, player, encounter)
            action(1, encounter,)
```
