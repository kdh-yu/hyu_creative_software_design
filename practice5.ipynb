#### What percent of sales occurred at each store type?

```python
# Calculate total weekly sales
sales_all = sales["weekly_sales"].sum()
```

```python
# Subset for type A stores, calculate total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()
sales_A
```

```python
sales[sales["type"] == "A"].weekly_sales.sum()
```

```python
# Subset for type B stores, calculate total weekly sales
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()
```

```python
# Subset for type C stores, calculate total weekly sales
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
```

```python
# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
## Better solution using .groupby()
# Group by type; calculate total weekly sales
sales_by_type = sales.groupby("type")["weekly_sales"].sum()
sales_by_type
```

```python
################################## TODO ##################################
# Get proportion for each type
# sales_propn_by_type = ?
sales_propn_by_type_2 = sales_by_type / sales_all
print(sales_propn_by_type_2)
```

```python
# From previous step
sales_by_type = sales.groupby("type")["weekly_sales"].sum()
```

```python
################################## TODO ##################################
# Group by type and is_holiday; calculate total weekly sales
# sales_propn_by_type_is_holiday = ?

sales_propn_by_type_is_holiday = sales.groupby(['type', 'is_holiday'])['weekly_sales'].sum()
sales_propn_by_type_is_holiday
```

```python
# Import NumPy with the alias np
import numpy as np
```

```python
# For each store type, aggregate weekly_sales: get min, max, mean, and median
sales_stats = sales.groupby("type")["weekly_sales"].agg([min, max, np.mean, np.median])

# Print sales_stats
print(sales_stats)
```

```python
################################## TODO ##################################
# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
# Use groupby
# unemp_fuel_stats = ?
unemp_fuel_stats = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg([min, max, np.mean, np.median])

# Print unemp_fuel_stats
print(unemp_fuel_stats)
```

```python
# For each store type, aggregate weekly_sales: get min, max, mean, and median
sales_stats = sales.groupby("type")["weekly_sales"].agg([np.min, np.max, np.mean, np.median])
sales.pivot_table(values='weekly_sales', index='type', aggfunc=[np.min, np.max, np.mean, np.median], fill_value = 0, margins=True)

# Print sales_stats
print(sales_stats)
```

```python
# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
unemp_fuel_stats = sales.groupby("type")[["unemployment", "fuel_price_usd_per_l"]].agg([np.min, np.max, np.mean, np.median])
# Print unemp_fuel_stats
print(unemp_fuel_stats)

################################## TODO ##################################
# Create a pivot table with the same results
pivot_unemp_fuel_stats = pd.pivot_table(unemp_fuel_stats, columns=['type'])
```

```python
print(pivot_unemp_fuel_stats)
```

```python
# For each pair of store type and is_holiday, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
################################## TODO ##################################
# Use group by
# unemp_fuel_stats_is_holiday = ?
unemp_fuel_stats_is_holiday = sales.groupby(['type', 'is_holiday'])[["unemployment", "fuel_price_usd_per_l"]].agg([np.min, np.max, np.mean, np.median])
print(unemp_fuel_stats_is_holiday)
################################## TODO ##################################
# Use pivot table
# pt_unemp_fuel_stats_is_holiday = ?
pt_unemp_fuel_stats_is_holiday = pd.pivot_table(unemp_fuel_stats_is_holiday, columns=['type', 'is_holiday'])
print(pt_unemp_fuel_stats_is_holiday)
```
